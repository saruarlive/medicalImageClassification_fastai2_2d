### Preprcessing, updating pandas dataframe,
Codes are partially modified from https://www.kaggle.com/jhonmgb/chexpert-inceptionresv2-clean, 

and from Satish https://github.com/skaliy/2d_medical_imaging_fastai_v2/blob/master/notebooks/CheXNet.ipynb

```python
u_one_features = ['Atelectasis', 'Edema']   #### -1 or 1
u_zero_features = ['Cardiomegaly', 'Consolidation', 'Pleural Effusion']   ### 0 or 1

##### Function for adding target features strings
def feature_string(row):
    feature_list = []
    for feature in u_one_features:
        if row[feature] in [-1,1]: 
            feature_list.append(feature)
            
    for feature in u_zero_features:
        if row[feature] == 1:
            feature_list.append(feature)
            
    return ';'.join(feature_list)

u_one_features.extend(u_zero_features)
def feature_blabel(row):   ###1 a group and -1, or NA, or ' ' or 0 another group 
    feature_list = []
    for feature in u_one_features:
        if row[feature] == 1:
            feature_list.append(feature)
                        
    return ';'.join(feature_list)
```

### Adding patient_id and study_session

```python
#df_train.Path.str.split('/',3,True)[3].str.split('_', 0, True)[0]
```

```python
df_train['patient_id'] = df_train.Path.str.split('/',3,True)[2]
df_train['study_view_id'] = df_train.Path.str.split('/',3,True)[3].str.split('_', 0, True)[0]

df_valid['patient_id'] = df_valid.Path.str.split('/',3,True)[2]
df_valid['study_view_id'] = df_valid.Path.str.split('/',3,True)[3].str.split('_', 0, True)[0]
```

```python
df_train.head()
```

```python
df_train['is_valid'] = False
df_valid['is_valid'] = True
```

```python
merged_df = pd.concat([df_train, df_valid])
merged_df.head()
```

```python
merged_df['feature_labels'] = merged_df.apply(feature_blabel,axis = 1).fillna('')
#merged_df['feature_labels'] =merged_df['feature_labels'] .apply(lambda x:x.split(";"))
merged_df.head()
```

```python
merged_df.loc[merged_df['No Finding'] == 1, 'feature_labels'] = 'No decision'
```

```python
nan_value = float("NaN")
merged_df.replace('', nan_value, inplace=True)
merged_df.dropna(subset=['feature_labels'], inplace=True)
merged_df.head()
```

```python
basePath
```

```python
dblock = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                   splitter=ColSplitter('is_valid'),
                   get_x=ColReader('Path', pref=basePath), 
                   get_y=ColReader('feature_labels', label_delim=';'),
                   item_tfms = Resize(512),
                   batch_tfms=[*aug_transforms(size=(224)), Normalize.from_stats(*imagenet_stats)])
```

```python
dls = dblock.dataloaders(merged_df, bs=64)
```

```python
dls.one_batch()[0].shape, dls.one_batch()[0].unique(), dls.one_batch()[0].dtype, dls.one_batch()[0].min(), dls.one_batch()[0].max(), dls.one_batch()[1].shape, dls.one_batch()[1].unique(), dls.one_batch()[1].dtype
```

```python
dls.show_batch(max_n=12)
```

```python
learn = cnn_learner(dls, resnet34, metrics=partial(accuracy_multi, thresh=0.2)).to_fp16()
```

```python
learn.model = nn.DataParallel(learn.model)
```

```python
lr_min,lr_steep = learn.lr_find()
lr_min,lr_steep
```

```python
learn.fine_tune(20, freeze_epochs=3)
```
